```python
import pandas as pd

# Read the first Excel file
file1 = 'C:\\Users\\Admin\\AIO\\Multivariate Time Series Forecasting (LSTM)\\excel_vnindex.xlsx'
df1 = pd.read_excel(file1)

# Read the second Excel file
file2 = 'C:\\Users\\Admin\\AIO\\Multivariate Time Series Forecasting (LSTM)\\bitcoin.xlsx'
df2 = pd.read_excel(file2)

# Convert date columns to datetime format if they are not already
df1['date time'] = pd.to_datetime(df1['date time'], format='%d/%m/%Y')
df2['Date'] = pd.to_datetime(df2['Date'], format='%d/%m/%Y')

# Merge DataFrames on the date columns
merged_df = pd.merge(df1, df2, left_on='date time', right_on='Date', how='outer')

# Print or manipulate merged DataFrame as needed
print(merged_df)
```
